```python
# Convert 'TransactionMonth' to datetime
df['TransactionMonth'] = pd.to_datetime(df['TransactionMonth'], errors='coerce')

# Filter relevant columns
claims_data = df[['TransactionMonth', 'TotalClaims', 'PolicyID']]
```

```python
# Group by month and aggregate claims
monthly_claims = claims_data.groupby(df['TransactionMonth'].dt.to_period('M')).agg(
    claim_count=('PolicyID', 'count'),         # Number of policies/transactions
    claim_frequency=('TotalClaims', lambda x: (x > 0).sum()),  # Number of months with claims
    total_claims=('TotalClaims', 'sum'),       # Total amount of claims
    avg_claim_severity=('TotalClaims', lambda x: x[x > 0].mean())  # Mean of non-zero claims
).reset_index()

# Convert back to datetime for plotting
monthly_claims['TransactionMonth'] = monthly_claims['TransactionMonth'].dt.to_timestamp()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(monthly_claims['TransactionMonth'], monthly_claims['claim_frequency'], marker='o')
plt.title("Monthly Claim Frequency Over Time")
plt.xlabel("Month")
plt.ylabel("Number of Claims")
plt.grid(True)
plt.tight_layout()
plt.show()
```

## 💼 In-Depth Interpretation: Loss Ratio Analysis of High-Claim Vehicle Models

This section investigates the profitability of the top 10 vehicle models with the highest average claim amounts by calculating their **Loss Ratio**, a key indicator used in insurance to evaluate financial performance.

### 🔍 What is Loss Ratio?

The **Loss Ratio** is calculated as:

\[
\text{Loss Ratio} = \frac{\text{Total Claims}}{\text{Total Premiums}}
\]

It quantifies how much the insurer pays in claims relative to the income received from premiums. A loss ratio:
- **Less than 1.0** indicates profitability (premiums exceed claims).
- **Greater than 1.0** implies loss-making (claims exceed premiums).
- **Significantly higher than 1.0** flags severe financial underperformance.

---

### 📊 Summary of Findings

After filtering the dataset to include only the 10 highest-claim vehicle models (based on average claim amount), the following results were observed:

- The **overall loss ratio** across these 10 models is **12.18**.
- Every single model had a loss ratio **above 7**, with the top offender (Hyundai H-1 2.5 CRDI Wagon A/T) showing a staggering **56.4**, meaning the insurer paid 56 times more in claims than it collected in premiums for that model.
- Models such as the **Toyota Land Cruiser FJ 4.0 V6**, **Volkswagen Polo 1.6 Comfortline**, and **Suzuki Ertiga 1.4 GA** also exhibit extreme unprofitability.

These results imply that even if a vehicle type appears frequently on the road or is otherwise popular, its **claim cost profile can make it an unsustainable business segment**.

---

### 🔎 Interpretation by Risk Level

1. **Very High-Risk Models (Loss Ratio > 20)**  
   - *Hyundai H-1* and *Toyota Land Cruiser* are causing catastrophic losses.
   - These models may be involved in severe accidents or are expensive to repair.
   - The pricing strategy for these models is likely **underestimating their risk**.

2. **High-Risk Models (Loss Ratio 8–15)**  
   - Models like the *Volkswagen Polo*, *Suzuki Ertiga*, and *Audi A4 Avant* fall into this category.
   - While less extreme than the top two, they still cost insurers far more than they're worth in premium.

3. **Moderately High-Risk (Loss Ratio 7–8)**  
   - Vehicles such as the *Mercedes-Benz C180 Coupe* still fall far short of profitability, but might be close enough for strategic adjustments.

---

### 💡 Business Implications

- **Product Design**: These models likely need targeted product adjustments — such as **higher excess amounts**, **co-payment options**, or **restricted coverage**.
- **Pricing Correction**: Premium pricing for these models is evidently misaligned with their risk level. A **recalibrated premium formula** based on historical claim severity is warranted.
- **Underwriting Policy**: Underwriters may need to apply **stricter approval conditions**, especially for older or modified versions of these vehicles.
- **Portfolio Optimization**: ACIS should assess whether to **reduce exposure** to these models or restructure the way these risks are underwritten and priced.

---

### 🚩 Warning Sign

If these models are **not adjusted or repriced**, they can continue to:
