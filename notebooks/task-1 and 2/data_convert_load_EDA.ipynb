##  In-Depth Interpretation: Loss Ratio Analysis of High-Claim Vehicle Models

This section investigates the profitability of the top 10 vehicle models with the highest average claim amounts by calculating their **Loss Ratio**, a key indicator used in insurance to evaluate financial performance.

### 🔍 What is Loss Ratio?

The **Loss Ratio** is calculated as:

\[
\text{Loss Ratio} = \frac{\text{Total Claims}}{\text{Total Premiums}}
\]

It quantifies how much the insurer pays in claims relative to the income received from premiums. A loss ratio:
- **Less than 1.0** indicates profitability (premiums exceed claims).
- **Greater than 1.0** implies loss-making (claims exceed premiums).
- **Significantly higher than 1.0** flags severe financial underperformance.


###  Summary of Findings

After filtering the dataset to include only the 10 highest-claim vehicle models (based on average claim amount), the following results were observed:

- The **overall loss ratio** across these 10 models is **12.18**.

- Models such as the **Toyota Land Cruiser FJ 4.0 V6**, **Volkswagen Polo 1.6 Comfortline**, and **Suzuki Ertiga 1.4 GA** also exhibit extreme unprofitability.

These results imply that even if a vehicle type appears frequently on the road or is otherwise popular, its **claim cost profile can make it an unsustainable business segment**.

## Imputation Strategy for `NewVehicle` Using Vehicle Age

### 🧾 Context

The `NewVehicle` column indicates whether a vehicle is **new** (i.e., purchased less than 6 months before the insurance transaction) or **used**. It contains:

- `More than 6 months`: 845,223 records
- `Less than 6 months`: 1,580 records
- **Missing (NaN)**: 153,295 records (~15.3%)

Due to the substantial proportion of missing values, we apply an informed imputation strategy using available vehicle information.



### Chosen Reference: `VehicleIntroDate` vs. `RegistrationYear`

- `VehicleIntroDate` contains the specific date the vehicle was introduced into the system.
- `RegistrationYear` is too coarse to reliably identify whether a vehicle is less than 6 months old — it lacks month-level precision.

 **Therefore, `VehicleIntroDate` is the preferred source** for estimating vehicle age.



### Method

1. **Convert** `VehicleIntroDate` and `TransactionMonth` to datetime format.
2. **Calculate vehicle age in months** using the difference between transaction and intro dates.
3. **Impute `NewVehicle`** values:
   - If `vehicle_age ≤ 6 months` → assign `"Less than 6 months"`
   - If `vehicle_age > 6 months` → assign `"More than 6 months"`
4. Optionally flag missing intro dates for downstream model awareness.






```python
df['VehicleIntroDate'] = pd.to_datetime(df['VehicleIntroDate'], errors='coerce')
df['TransactionMonth'] = pd.to_datetime(df['TransactionMonth'])

df['vehicle_age_days'] = (df['TransactionMonth'] - df['VehicleIntroDate']).dt.days
df['vehicle_age_months'] = df['vehicle_age_days'] / 30.44
```

```python
# Impute NewVehicle
df.loc[df['NewVehicle'].isnull() & (df['vehicle_age_months'] <= 6), 'NewVehicle'] = 'Less than 6 months'
df.loc[df['NewVehicle'].isnull() & (df['vehicle_age_months'] > 6), 'NewVehicle'] = 'More than 6 months'

```

## Handling Missing Values in Demographic Columns

To maintain the completeness of customer-related features without introducing bias or losing significant data, missing values in demographic columns were imputed based on context and data distribution.

Bank: Missing bank information may indicate non-disclosure or alternative payment methods. "Unknown" captures this explicitly without removing data.

AccountType: The distribution is usually skewed toward one dominant class — filling with the mode reduces distortion.

MaritalStatus and Gender: These columns have very low missingness. Imputing with the mode avoids unnecessary complexity and preserves the full sample size for modeling.


```python
# Impute categorical demographic fields safely
df['Bank'] = df['Bank'].fillna('Unknown')
df['AccountType'] = df['AccountType'].fillna(df['AccountType'].mode()[0])
df['MaritalStatus'] = df['MaritalStatus'].fillna(df['MaritalStatus'].mode()[0])
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
```
